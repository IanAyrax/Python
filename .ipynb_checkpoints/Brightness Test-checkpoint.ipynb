```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math

#load image and show it
normal1 = './Dataset/Normal/1.JPG'
sedang1 = "./Dataset/sedang/2.JPG"
sedang2 = "./Dataset/sedang/IMG_0358.JPG"
tinggi1 = "./Dataset/tinggi/1.JPG"

img_normal1 = cv2.imread(normal1)
img1 = cv2.imread(sedang1)
img2 = cv2.imread(sedang2)
```

```python
import imageio
import numpy as np

f = imageio.imread(sedang2, as_gray=True)

def img_estim(img, thrshld):
    print("Brightness : " + str(np.mean(img)))
    is_light = np.mean(img) > thrshld
    return 'light' if is_light else 'dark'

print(img_estim(f, 127))
```

```python
import os

direktori = "./Dataset/"
for count, data in enumerate(os.listdir(direktori + "tinggi")) :
    f = imageio.imread(direktori + "tinggi" + "/" + data, as_gray=True)
    print(img_estim(f, 127))
```

```python
direktori = "./Dataset/"
for count, data in enumerate(os.listdir(direktori + "sedang")) :
    f = imageio.imread(direktori + "sedang" + "/" + data, as_gray=True)
    print(img_estim(f, 127))
```
