```python
def adjust_gamma(image, gamma=1.0):
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
    for i in np.arange(0, 256)]).astype("uint8")

    return cv2.LUT(image, table)

def decrease_brightness(image, value=30):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)
    
    lim = 255 - value
    v[v > lim] = 255
    v[v <= lim] += value
    
    final_hsv = cv2.merge((h, s, v))
    img = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
    
    return img

#img_pure = img_normal.copy()
#img_test = img_normal.copy()
#img_result = img_normal.copy()

#img_pure = img_sedang.copy()
#img_test = img_sedang.copy()
#img_result = img_sedang.copy()

img_pure = img_tinggi.copy()
img_result = img_tinggi.copy()
img_test = img_tinggi.copy()

plt.imshow(img_test)
plt.show()

gamma = 0.001                             # change the value here to get different result
gamma_img = adjust_gamma(img_test, gamma=gamma)

plt.imshow(gamma_img)
plt.show()

brightness = decrease_brightness(img_test, value=20)
plt.imshow(brightness)
plt.show()
```

```python
pixels = np.argwhere(otsu_img == 255)
i = 0
x_max = 0
x_min = 0
y_max = 0
y_min = 0
for pixel in pixels :
    if i == 0 :
        y_max = pixel[0]
        y_min = pixel[0]
        x_max = pixel[1]
        x_min = pixel[1]
    i = i + 1;
    if x_max < pixel[1] :
        x_max = pixel[1]
        
    if x_min > pixel[1] :
        x_min = pixel[1]
        
    if y_max < pixel[0] :
        y_max = pixel[0]
        
    if y_min > pixel[0] :
        y_min = pixel[0]
        
center_x = int(x_min + (x_max - x_min) / 2)
center_y = int(y_min + (y_max - y_min) / 2)

eye_center = [center_x, center_y]

print(str(x_min) + ", " + str(y_min))
print(str(x_max) + ", " + str(y_max))
print(eye_center)

plt.imshow(otsu_img)
plt.show()

cv2.circle(img_result, (center_x,center_y), 50, (0, 255, 0), -1)
cv2.circle(img_result, (x_min, y_min), 50, (255, 0, 0), -1)
cv2.circle(img_result, (x_max, y_max), 50, (0, 0, 255), -1)
plt.imshow(img_result)
plt.show()
```

```python
param1 = 90
#param2 = 800
param2 = 1000
maxRadius = 560
#maxRadius = 780
#minRadius = 450
#minRadius = 340
minRadius = 250
#minRadius = 200

#eye_center = [1840, 1010] #Normal
#eye_center = [1700, 1350] #Tinggi

#hough_circle = cv2.HoughCircles(img_test, cv2.HOUGH_GRADIENT, 1.3, 800)
#hough_circle = cv2.HoughCircles(img_test, cv2.HOUGH_GRADIENT, 4, 800)
#hough_circle = cv2.HoughCircles(img_test, cv2.HOUGH_GRADIENT, 7, 20, param1=param1, param2=param2, minRadius=minRadius, maxRadius=maxRadius)
hough_circle = cv2.HoughCircles(img_test, cv2.HOUGH_GRADIENT, 4, 800, param1=param1, param2=param2, minRadius=minRadius, maxRadius=maxRadius)

i = 0
choosen_x = 0
choosen_y = 0
choosen_radius = 0

if hough_circle is not None :
    hough_circle = np.round(hough_circle[0, :]).astype("int")
    for (x, y, radius) in hough_circle :
        i = i + 1
        if i == 1 :
            choosen_x = x
            choosen_y = y
            choosen_radius = radius
        
        distance_a  = math.sqrt((choosen_x - eye_center[0]) * (choosen_x - eye_center[0]) + (choosen_y - eye_center[1]) * (choosen_y - eye_center[1]))
        distance_b  = math.sqrt((x - eye_center[0]) * (x - eye_center[0]) + (y - eye_center[1]) * (y - eye_center[1]))
        
        if distance_a > distance_b :
            choosen_x = x
            choosen_y = y
            choosen_radius = radius
            
        #cv2.circle(img_result, (x,y), radius, (255, 0, 0), 4)
    cv2.circle(img_result, (choosen_x,choosen_y), choosen_radius, (255, 0, 0), 4)
    
print(i)
print("Radius : " + str(choosen_radius))
print("X, Y : " + str(choosen_x) + ", " + str(choosen_y))
plt.imshow(img_result)
plt.show()
```

```python
new_radius = int(choosen_radius * 0.7)
padding = 280
image = img_pure[choosen_y - choosen_radius - padding:choosen_y + choosen_radius + padding, choosen_x - choosen_radius - padding:choosen_x + choosen_radius + padding]

plt.imshow(image)
plt.show()

image = cv2.GaussianBlur(image, (7,7), 1)
#height = int(choosen_radius - new_radius)
width = int(choosen_radius * 2 * np.pi)
height = new_radius
r_in = new_radius
r_out = choosen_radius - 50

thetas = np.arange(0, 2 * np.pi, 2 * np.pi / width)  # Theta values
#r_out = r_in + r_out
# Create empty flatten image
flat = np.zeros((height,width, 3), np.uint8)
circle_x = int(image.shape[0] / 2)
circle_y = int(image.shape[1] / 2)

for i in range(width):
    for j in range(height):
        theta = thetas[i]  # value of theta coordinate
        r_pro = j / height  # value of r coordinate(normalized)

        # get coordinate of boundaries
        Xi = circle_x + r_in * np.cos(theta)
        Yi = circle_y + r_in * np.sin(theta)

        Xo = circle_x + r_out * np.cos(theta)
        Yo = circle_y + r_out * np.sin(theta)

        # the matched cartesian coordinates for the polar coordinates
        Xc = (1 - r_pro) * Xi + r_pro * Xo
        Yc = (1 - r_pro) * Yi + r_pro * Yo

        if Xc >= image.shape[0] :
            Xc = image.shape[1] - 1

        if Yc >= image.shape[1] :
            Yc = image.shape[1] - 1
        
        color = image[int(Xc)][int(Yc)]  # color of the pixel
        flat[j][i] = color

plt.imshow(flat)
plt.show()        

#gray_flat = cv2.cvtColor(flat, cv2.COLOR_BGR2GRAY)
#flat[np.where((flat!=[255,255,255]).all(axis = 2))] = [0,0,0]
flat[np.where((flat < [99, 110, 140]).all(axis = 2))] = [0,0,0]

gray_flat = cv2.cvtColor(flat, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat)
plt.show()

otsu_threshold, image_result = cv2.threshold(gray_flat, 0, 255, cv2.THRESH_OTSU)
histr = cv2.calcHist([image_result],[0],None,[256],[0,256])

print(otsu_threshold)
plt.imshow(image_result)
plt.show()

plt.plot(histr)
plt.show()
```
